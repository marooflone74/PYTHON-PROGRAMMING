## **1. What is File I/O?**
File I/O (Input/Output) refers to how a program interacts with external files, such as **reading from** or **writing to** them. In Python, we use built-in functions like `open()` to work with files.

Examples of File I/O operations:
- Reading a text file.
- Writing data to a file.
- Working with structured data (JSON, CSV, etc.).


## **2. Why Do We Need File Handling?**
File handling is useful when we need to:
- Store and retrieve **persistent data**.
- Share data between different programs.
- Work with **large datasets** efficiently.

**Example Use Case:**
- Saving **user preferences** in an application.
- Storing **logs** for analysis.


## **3. Opening and Closing Files in Python**
To work with files in Python, we use the `open()` function.

**Syntax:**
```python
file = open('filename.txt', 'mode')
# Perform operations on the file
file.close()
```
Always close the file after opening it to free up system resources!


````python
# Opening a file and closing it properly
file = open('example.txt', 'w')
file.write('Hello, this is a sample file!')
file.close()
````

##  **4. Reading from and Writing to Files**
We can use different modes while working with files:
- `'r'` → Read mode (default)
- `'w'` → Write mode (creates a new file or overwrites an existing one)
- `'a'` → Append mode (adds data without deleting existing content)
- `'r+'` → Read and write mode

 **Example:** Writing and reading a file in Python.


###  **Examples of Different File Modes in Python**
Below are examples demonstrating the different file modes available in Python:


####  **Read Mode (`'r'`)** - Opens a file for reading. The file must exist.

````python
try:
    with open('example.txt', 'r') as file:
        content = file.read()
        print('File Content:', content)
except FileNotFoundError:
    print('Error: The file does not exist!')
````

####  **Write Mode (`'w'`)** - Creates a new file or overwrites an existing one.

````python
with open('example.txt', 'w') as file:
    file.write('This is a new file created using write mode.')
print('File written successfully.')
````

####  **Append Mode (`'a'`)** - Adds data to a file without deleting existing content.

````python
with open('example.txt', 'a') as file:
    file.write('\nThis is a new line added using append mode.')
print('New content appended successfully.')
````

####  **Read & Write Mode (`'r+'`)** - Opens the file for both reading and writing.

````python
with open('example.txt', 'r+') as file:
    content = file.read()
    print('Current Content:', content)
    file.write('\nAdding new content in read+write mode.')
print('Content updated successfully.')
````

````python
# Writing to a file
with open('example.txt', 'w') as file:
    file.write('This is a new line.')

# Reading from a file
with open('example.txt', 'r') as file:
    content = file.read()
    print(content)
````

##  **5. Working with JSON Files in Python**
JSON (JavaScript Object Notation) is a popular format for **storing and exchanging data**.

Python provides a built-in `json` module to handle JSON files easily.

 **Example:** Storing and loading data using JSON files.


````python
import json

# Sample dictionary
data = {'name': 'Haris', 'age': 25, 'city': 'Lahore'}

# Writing JSON data to a file
with open('data.json', 'w') as file:
    json.dump(data, file)

# Reading JSON data from a file
with open('data.json', 'r') as file:
    loaded_data = json.load(file)
    print(loaded_data)
````

##  **6. Handling File Exceptions**
Sometimes, errors occur while working with files. For example:
- The file does not exist.
- The program lacks permission to read/write.
- The file is already in use by another program.

 **Use `try-except` to handle file-related exceptions.**


````python
try:
    with open('non_existent_file.txt', 'r') as file:
        content = file.read()
        print(content)
except FileNotFoundError:
    print('Error: The file does not exist!')
````

##  **7. Best Practices in File Handling**
✔️ Always close the file after opening it (or use `with open()` for automatic closing).

✔️ Use `try-except` to handle file-related errors gracefully.

✔️ Avoid hardcoding file paths—use `os.path.join()` for compatibility.

✔️ Use `json.dump()` and `json.load()` for structured data instead of manually formatting text files.

````python
# 1. write a program that count a line in aa file
def count_lines(filename):
    try:
        with open(filename, 'r') as file:
            lines = file.readlines()
            return len(lines)
    except FileNotFoundError:
        print('Error: The file does not exist!')
        return 0    
line_count = count_lines('example.txt')
print(f'The number of lines in the file is: {line_count}')
````

````python
# write a program that count a word in aa file
def count_words(filename):
    try:
        with open(filename, 'r') as file:
            content = file.read()  # Read the entire content of the file as a string
            words = content.split() # Split the content into words based on whitespace
            return len(words)
    except FileNotFoundError:
        print('Error: The file does not exist!')
        return 0

word_count = count_words('example.txt')
print(f'The number of words in the file is: {word_count}')
````

````python
# 2️⃣ **Check If a File Exists** - Write a program that asks the user for a filename and checks whether it exists before opening it.

import os
filename = input('enter the filename: ')
if os.path.isfile(filename):
    with open(filename, 'r') as file:
        content = file.read()
        print(content)
else:
    print('Error: The file does not exist!')
    
````

````python
#3️⃣ **Reverse File Content** - Write a program that reads a file's content and writes a reversed version of it into a 
#new file.
def reverse_file_content(input_file, output_file):
    try:
        with open(input_file, 'r') as file:
            content = file.read()
            reversed_content = content[::-1]  # Reverse the content
        with open(output_file, 'w') as file:
            file.write(reversed_content)
        print(f'Reversed content has been written to {output_file}')
    except FileNotFoundError:
        print('Error: The input file does not exist!')

input_file = 'example.txt'
output_file = 'reversed_example.txt'
reverse_file_content(input_file, output_file)


            
````

````python
#4️⃣ **Copy File Content** - Write a program that copies the content of one file into another file.
def copy_file_content(source_file, destination_file):
    try:
        with open(source_file, 'r') as file:
            content = file.read()
        with open(destination_file, 'w') as file:
            file.write(content)
        print(f'Content has been copied from {source_file} to {destination_file}')
    except FileNotFoundError:
        print('Error: The source file does not exist!')
source_file = 'example.txt'
destination_file = 'copied_example.txt'
copy_file_content(source_file, destination_file)
````

````python
#5️⃣ **Merge Multiple Files** - Write a program that merges the content of two files into a single file.
def merge_files(file1, file2, merged_file):
    try:
        with open(file1, 'r') as f1, open(file2, 'r') as f2:
            content1 = f1.read()
            content2 = f2.read()
        with open(merged_file, 'w') as f3:
            f3.write(content1 + content2)
        print(f'Files {file1} and {file2} have been merged into {merged_file}')
    except FileNotFoundError:
        print('Error: One or more of the specified files do not exist!')
file1 = 'example.txt'
file2 = 'reversed_example.txt'
merged_file = 'merged_example.txt'
merge_files(file1, file2, merged_file)
````

````python
#6️⃣ **Log File with Timestamp** - Create a logging system where every time the program runs, it appends a new log entry with a timestamp into a log file.
# explanation of code: this code defines a function called `log_message` that takes a message and an optional log file name (default is 'log.txt'). It gets the current date and time, formats it as a string, and creates a log entry that includes the timestamp and the message. The function then opens the specified log file in append mode, writes the log entry to the file, and prints a confirmation message that the log entry has been added. 
import datetime
def log_message(message, log_file='log.txt'):# Define a function that takes a message and an optional log file name (default is 'log.txt')
    timestamp = datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')# Get the current date and time, and format it as a string
    log_entry = f'[{timestamp}] {message}\n'# Create a log entry string that includes the timestamp and the message
    with open(log_file, 'a') as file:# Open the log file in append mode to add new entries without overwriting existing ones
        file.write(log_entry)# Write the log entry to the file
    print(f'Log entry added: {log_entry.strip()}')# Print a confirmation message that the log entry has been added, stripping any extra whitespace from the log entry
log_message('This is a log entry.')

````
